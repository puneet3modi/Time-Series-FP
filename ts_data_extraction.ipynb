```python
import pandas as pd
```

```python
path = '/Users/pandita/Documents/MER_T11_01.csv'
ce = pd.read_csv(path)

# Convert YYYYMM to a datetime format assuming the day is always the first of the month
ce['Date'] = pd.to_datetime(ce['YYYYMM'], format='%Y%m', errors='coerce')

# Drop rows where date conversion failed (if any)
ce = ce.dropna(subset=['Date'])

# Create the pivot table
ce = ce.pivot_table(values='Value', index='Date', columns='Description', aggfunc='first')

# Display the first few rows of the pivot table
ce.reset_index(inplace=True)
ce.head()
```

```python
ep['Month'] = pd.to_datetime(ep['Month'], format='%Y%m', errors='coerce')
ec['Month'] = pd.to_datetime(ec['Month'], format='%Y%m', errors='coerce')
```

```python
from fredapi import Fred

# Replace 'your_api_key_here' with your actual FRED API key
fred = Fred(api_key='b73c11e70992c0501f8748360e192763')

# Economic Variables
gdp = fred.get_series('GDP',frequency = 'q')  # Gross Domestic Product (GDP)
employment_rate = fred.get_series('EMRATIO',frequency = 'q')  # Civilian Employment-Population Ratio
industrial_output = fred.get_series('INDPRO',frequency = 'q')  # Industrial Production
```

```python
# concat the economic variables into a single DataFrame , for date range : 1973-01-01 to 2017-01-01
econ_df = pd.concat([gdp, employment_rate, industrial_output], axis=1)
econ_df.columns = ['GDP', 'Employment Rate', 'Industrial Output']
```

```python
econ_df = econ_df.resample('MS').mean()
econ_df = econ_df.interpolate(method='linear')

econ_df.reset_index(inplace=True)
econ_df = econ_df.rename(columns={'index': 'Date'})
econ_df
```

```python
# Merge the two DataFrames
merged = pd.merge(merged, econ_df, on='Date')
```
